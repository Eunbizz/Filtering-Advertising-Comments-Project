```python
df3.rename(columns={'content':'text'}, inplace=True)
df4.rename(columns={'content':'text'}, inplace=True)
```

```python
data=pd.concat([df1,df2,df3,df4],ignore_index=True)
data
```

```python
data['label'].value_counts()
```

```python
#숫자 제거
p = re.compile("[0-9]+")
for i in range (0,20947):
  data['text'][i]=p.sub(" ",data['text'][i])
```

```python
#문장부호,특수문자 제거
p = re.compile("\W+")
for i in range (0,20947):
  data['text'][i]=p.sub(" ",data['text'][i])

#문장부호,특수문자 제거
p = re.compile("_")
for i in range (0,20947):
  data['text'][i]=p.sub(" ",data['text'][i])
```
