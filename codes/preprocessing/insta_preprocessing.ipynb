```python
import pandas as pd
import numpy as np 
import warnings
warnings.filterwarnings(action='ignore')
```

#### csv 파일 호출

```python
data = pd.read_csv('label0.csv')
data
```

```python
true_review = pd.read_csv('솔직리뷰_pre.csv')
true_review
```

#### 결측값 제거

```python
data = data.dropna()
true_review = true_review.dropna()
```

```python
data.reset_index(inplace=True)
true_review.reset_index(inplace=True)
```

```python
data = data.drop(labels = ['index'],axis=1)
true_review = true_review.drop(labels = ['index'],axis=1)
```

```python
data
```

```python
true_review
```

#### 뒤에 날짜 제거 코드(사용 x) 

```python
print(data['content'][0][-2:])
```

```python
for i in range(len(data)):
    data['content'][i] = data['content'][i][:-3]
```

```python
for i in range(len(data)):
    s = data['content'][i][-1]
    if type(s) == type(1):
        data['content'][i] = data['content'][i][:-2]
```

```python
data
```

#### 인스타그램 아이디 제거 

```python
def decision(s):
    if not s.isalpha():
        if not s.isdigit():
            if s != '_':
                if s != '.':
                    return True
    return False
```

```python
for i in range(len(data)):
    for j in range(len(data['content'][i])):
        if decision(data['content'][i][j]):
            data['content'][i] = data['content'][i][j:]
            break
```

```python
data
```

#### 본문 내 해시태그 제거

```python
for i in range(len(data)):
    data['content'][i] = data['content'][i].replace('#','')

data
```

#### 단어 필터링

```python
# 필터링 단어들
stop_word_list=[
                '#인스타그램광고',
                '#인스타그램홍보문의',
                '문의',
                '방판',
                '제공',
                '문의환영',
                '제품문의',
                '무자본',
                '카운셀러',
                '주문 감사',
                '단체예약',
                '클릭',
                '링크',
                '행사',
                '공구',
                '공동구매',
                '부업',
                '고객',
                '오픈기념',
                '공지',
                '제품제공',
                '협찬문의',
                '#대외활동',
                '#방문판매',
                '문의는',
                '원장',
                '#모든문의'
]
```

```python
# data 
for i,stl in enumerate(stop_word_list):
    data=data[~data['content'].str.contains(stl)]
data   
```

```python
data.reset_index(inplace=True)
data.drop('index',axis=1, inplace=True)
data
```

```python
# true_review
for i,stl in enumerate(stop_word_list):
    true_review=true_review[~true_review['content'].str.contains(stl)]
true_review   
```

```python
true_review.reset_index(inplace=True)
true_review.drop('index',axis=1, inplace=True)
true_review
```

#### 데이터 통일

```python
data = pd.concat([data, true_review], axis=0)
data
```

```python
data.to_csv('label0_insta.csv', encoding = 'utf-8-sig', index = False)
```
